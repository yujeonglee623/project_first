```python
import torch

# CUDA 사용 가능 여부 확인
print("CUDA 사용 가능 여부:", torch.cuda.is_available())

# GPU 개수 확인
print("GPU 개수:", torch.cuda.device_count())

# 현재 GPU 이름 확인
if torch.cuda.is_available():
    print("현재 GPU 이름:", torch.cuda.get_device_name(0))
else:
    print("GPU를 찾을 수 없습니다.")

# 간단한 텐서 연산 테스트
if torch.cuda.is_available():
    x = torch.rand(3, 3).to("cuda")
    y = torch.rand(3, 3).to("cuda")
    z = x + y
    print("GPU에서 연산 성공! 결과 텐서:")
    print(z)
```

```python
import torch, time

# CPU 연산
x_cpu = torch.rand(10000, 10000)
start = time.time()
_ = x_cpu @ x_cpu
print("CPU 연산 시간:", time.time() - start)

# GPU 연산
if torch.cuda.is_available():
    x_gpu = x_cpu.to("cuda")
    torch.cuda.synchronize()
    start = time.time()
    _ = x_gpu @ x_gpu
    torch.cuda.synchronize()
    print("GPU 연산 시간:", time.time() - start)
```
